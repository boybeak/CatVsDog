```python
# Create symlink for pics

# 引入模块
import os

def mkdir(path):
 
    # 去除首位空格
    path=path.strip()
    # 去除尾部 \ 符号
    path=path.rstrip("\\")
 
    # 判断路径是否存在
    # 存在     True
    # 不存在   False
    isExists=os.path.exists(path)
 
    # 判断结果
    if not isExists:
        # 如果不存在则创建目录
        print (path+' 创建成功')
        # 创建目录操作函数
        os.makedirs(path)
        return True
    else:
        # 如果目录存在则不创建，并提示目录已存在
        print (path+' 目录已存在')
        return False

parent = os.getcwd()
remote_train = 'remote_sub_train/' if is_debug else 'remote_train/'
remote_train_dir_abs = parent + '/' + remote_train
mkdir(remote_train_dir_abs)
train_dir_abs = parent + '/' + train_folder

# 引入模块
import os

file_list = os.listdir(train_folder)
for f in file_list:
    dst = remote_train_dir_abs + f
    if (not os.path.exists(dst)):
        os.symlink(train_dir_abs + f, dst, target_is_directory=False)
```
